### 1. Generate a tree graph that represents flipping a coin 4 times,  let A be the event “the first outcome is tails” , B the event “ the second outcome is head” and C the event “the third outcome is tails” calculate 

p(AUBUC) = P (A U B U C) = P(A) + P(B) + P(C) - P(A ∩ B) - P(A ∩ C) - P(B ∩ C) + P(A ∩ B ∩ C)

```r
library(igraph)

g <- graph.tree(n = 2^5-1, children = 2)

coin <- c("H", "T")
n.labels <- c("", replicate(15, coin))
e.labels <- c("1/2")
e2.labels <- replicate(30, e.labels)
```

```r
V(g)$color <- "#C4D8E2"
coords <- layout_(g, as_tree())
coord2 = matrix(c(-coords[,2],-coords[,1]),ncol = 2)
```

```r
plot(g,
     layout = coord2,           # draw graph as tree
     vertex.size = 20,                  # node size
     vertex.color = V(g)$color,          # node color
     vertex.label = n.labels,        # node labels
     vertex.label.cex = 1,             # node label size
     vertex.label.family = "Helvetica", # node label family
     vertex.label.font = 2,             # node label type (bold)
     vertex.label.color = '#000000',    # node label size
     edge.label = e2.labels,          # edge labels
     edge.label.cex = .7,               # edge label size
     edge.label.family = "Helvetica",   # edge label family
     edge.label.font = 1,               # edge label font type (bold)
     edge.label.color = '#000000',      # edge label color
     edge.arrow.size = 0.2,              # arrow size
     edge.arrow.width = 1              # arrow width
)
```

```r
"Probability Calculation for coin toss using p(AUBUC) 
= P (A U B U C) = P(A) + P(B) + P(C) - P(A ∩ B) - P(A ∩ C) - P(B ∩ C) + P(A ∩ B ∩ C)"
A <- 1/2
B <- 1/2
C <- 1/2
AB <- 1/4
AC <- 1/4
BC <- 1/4
ABC <- 1/8
P <- A + B + C -AB - AC - BC + ABC #calculation
'probablilty for p(AuBuC)='
P
```

### 2. From the Dataset Diabetes, construct contingency tables for the following variable combinations:

A: location Vs gender
B: Gender Vs frame
C: Gender Vs Age (Convert age to an discrete ordinal variable with three categories)
D: Cholesterol Vs Age (Convert age and cholesterol to an discrete ordinal variable with three categories)

calculate the joint and marginal probabilities, and from the above contingency tables choose 5 conditional probability examples with the probabilities calculations and one or two sentences explaining the results.

```r
library("ggplot2")
library("dplyr")
library("reshape2")
library("knitr")
```

```r
#Read diabetes.csv file
diabetes = read.csv(file = "diabetes.csv")
head(diabetes)
```

```r
'A = Location and gender'
diabetes.location.gender.df <-
  diabetes %>%
  group_by(location, gender) %>%
  #filter(Sex != "Both") %>%
  summarize(n = n())
diabetes.location.gender.df %>%
  dcast(location ~ gender, value.nar = "n") %>%
  kable(align = "l", format = "markdown",
        table.attr='class="table table-striped table-hover"')

'B = Gender and frame'
diabetes.gender.frame.df <-
  diabetes %>%
  group_by(gender, frame) %>%
  summarize(n = n())
diabetes.gender.frame.df %>%
  dcast(gender ~ frame, value.nar = "n") %>%
  kable(align = "l", format = "markdown",
        table.attr='class="table table-striped table-hover"')

#group age into 3 groups
diabetes$age_group <- with(diabetes, cut(age, breaks=3, labels=1:3))
#group cholesterol into 3 groups
diabetes$chol_group <- with(diabetes, cut(chol, breaks=3, labels=1:3))

'C = Gender and age groups'
diabetes.gender.age.df <-
  diabetes %>%
  group_by(gender, age_group) %>%
  summarize(n = n())
diabetes.gender.age.df %>%
  dcast(gender ~ age_group, value.nar = "n") %>%
  kable(align = "l", format = "markdown",
        table.attr='class="table table-striped table-hover"')

'D = Cholesterol and age group'
diabetes.chol.age.df <-
  diabetes %>%
  group_by(chol_group, age_group) %>%
  summarize(n = n())
diabetes.chol.age.df %>%
  dcast(chol_group ~ age_group, value.nar = "n") %>%
  kable(align = "l", format = "markdown",
        table.attr='class="table table-striped table-hover"')
```

```r
#Creating marginal table for all A, B, C, D

diabetes.location.gender.prop.df <-
  diabetes.location.gender.df %>%
  ungroup() %>%
  mutate(prop = n/ sum(n))

diabetes.location.gender.df %>%
  dcast(location ~ gender, value.nar = "prop") %>%
  kable(align = "l", format = "markdown",
        table.attr='class="table table-striped table-hover"')

diabetes.gender.frame.prop.df <-
  diabetes.gender.frame.df %>%
  ungroup() %>%
  mutate(prop = n/ sum(n))

diabetes.gender.frame.df %>%
  dcast(gender ~ frame, value.nar = "prop") %>%
  kable(align = "l", format = "markdown",
        table.attr='class="table table-striped table-hover"')

diabetes.gender.age.prop.df <-
  diabetes.gender.age.df %>%
  ungroup() %>%
  mutate(prop = n/ sum(n))

diabetes.gender.age.df %>%
  dcast(gender ~ age_group, value.nar = "prop") %>%
  kable(align = "l", format = "markdown",
        table.attr='class="table table-striped table-hover"')

diabetes.chol.age.prop.df <-
  diabetes.chol.age.df %>%
  ungroup() %>%
  mutate(prop = n/ sum(n))

diabetes.chol.age.df %>%
  dcast(chol_group ~ age_group, value.nar = "prop") %>%
  kable(align = "l", format = "markdown",
        table.attr='class="table table-striped table-hover"')
```

```r
#Probability for A, B, C, D
'A'
location.marginal.df <- 
  diabetes.location.gender.prop.df %>%
  group_by(location) %>%
  summarize(marginal = sum(prop))

gender.marginal.df <- 
  diabetes.location.gender.prop.df %>%
  group_by(gender) %>%
  summarize(marginal = sum(prop))

diabetes.location.gender.prop.df %>%
  dcast(location ~ gender, value.var = "prop") %>%
  left_join(location.marginal.df, by = "location") %>%
  bind_rows(
    gender.marginal.df %>%
      mutate(location = "marginal") %>%
      dcast(gender ~ location, value.var = "marginal")
  ) %>%
  kable(align = "l", format = "markdown",
        table.attr = 'class="table table-striped table-hover"')

'B'
gender.marginal.df <- 
  diabetes.gender.frame.prop.df %>%
  group_by(gender) %>%
  summarize(marginal = sum(prop))

frame.marginal.df <- 
  diabetes.gender.frame.prop.df %>%
  group_by(frame) %>%
  summarize(marginal = sum(prop))

diabetes.gender.frame.prop.df %>%
  dcast(gender ~ frame, value.var = "prop") %>%
  left_join(gender.marginal.df, by = "gender") %>%
  bind_rows(
    frame.marginal.df %>%
      mutate(gender = "marginal") %>%
      dcast(gender ~ frame, value.var = "marginal")
  ) %>%
  kable(align = "l", format = "markdown",
        table.attr = 'class="table table-striped table-hover"')
'C'
gender.marginal.df <- 
  diabetes.gender.age.prop.df %>%
  group_by(gender) %>%
  summarize(marginal = sum(prop))

frame.marginal.df <- 
  diabetes.gender.age.prop.df %>%
  group_by(age_group) %>%
  summarize(marginal = sum(prop))

diabetes.gender.age.prop.df %>%
  dcast(gender ~ age_group, value.var = "prop") %>%
  left_join(gender.marginal.df, by = "gender") %>%
  bind_rows(
    frame.marginal.df %>%
      mutate(gender = "marginal") %>%
      dcast(gender ~ age_group, value.var = "marginal")
  ) %>%
  kable(align = "l", format = "markdown",
        table.attr = 'class="table table-striped table-hover"')
'D'
chol.marginal.df <- 
  diabetes.chol.age.prop.df %>%
  group_by(chol_group) %>%
  summarize(marginal = sum(prop))

age.marginal.df <- 
  diabetes.chol.age.prop.df %>%
  group_by(age_group) %>%
  summarize(marginal = sum(prop))

diabetes.chol.age.prop.df %>%
  dcast(chol_group ~ age_group, value.var = "prop") %>%
  left_join(chol.marginal.df, by = "chol_group") %>%
  bind_rows(
    age.marginal.df %>%
      mutate(chol_group = "marginal") %>%
      dcast(chol_group ~ age_group, value.var = "marginal")
  ) %>%
  kable(align = "l", format = "markdown",
        table.attr = 'class="table table-striped table-hover"')
```

```r
#Join and marginal probabillity
"
I created 5 different probability conditions: 
    #1 is the probability of male in the first age group
    #2 is the probability of cholesterol in group 3 within age group 1
    #3 is the probability of location Buckingham with female gender
    #4 is the probability of female with small frame
    #5 is the probability of female in the second age group
"

'Condition #1'
joint.prob <- 
  diabetes.gender.age.prop.df %>%
  filter(gender == "male", age_group == 1) %>%
  .$prop

marg.prob <- 
  gender.marginal.df %>%
  filter(gender == "male") %>%
  .$marginal

cond.prob <- joint.prob / marg.prob
cond.prob

'Condition #2'
joint.prob <- 
  diabetes.chol.age.prop.df %>%
  filter(chol_group == 3, age_group == 1) %>%
  .$prop

marg.prob <- 
  chol.marginal.df %>%
  filter(chol_group == "3") %>%
  .$marginal

cond.prob <- joint.prob / marg.prob
cond.prob

'Condition #3'
joint.prob <- 
  diabetes.location.gender.prop.df %>%
  filter(location == "Buckingham", gender == "female") %>%
  .$prop

marg.prob <- 
  location.marginal.df %>%
  filter(location == "Buckingham") %>%
  .$marginal

cond.prob <- joint.prob / marg.prob
cond.prob

'Condition #4'
joint.prob <- 
  diabetes.gender.frame.prop.df %>%
  filter(gender == "female", frame == "small") %>%
  .$prop

marg.prob <- 
  gender.marginal.df %>%
  filter(gender == "female") %>%
  .$marginal

cond.prob <- joint.prob / marg.prob
cond.prob

'Condition #5'
joint.prob <- 
  diabetes.gender.age.prop.df %>%
  filter(gender == "female", age_group == 2) %>%
  .$prop

marg.prob <- 
  gender.marginal.df %>%
  filter(gender == "female") %>%
  .$marginal

cond.prob <- joint.prob / marg.prob
cond.prob
```

### 3. Baye’s Rule

Write a function in R that allows you to use the Baye’s rule for multiple events.

Use it to calculate the following problem:

Different isoforms of fundamental Hormones such as testosterone are relatively associated to behavioral differences in humans such as extreme aggression.  In a study where 75 % of the participants had the Isoform A, and 25 % had the isoform B. 54 people exhibited extreme aggression out 95  that had Isoform A, and 34 people with extreme aggression out of 90 that had the isoform B. 

Calculate the probability that someone with the isoform A exhibits extreme aggression. 
Calculate the probability that someone with the isoform B exhibits extreme aggression. 

```r
#Baye's Rule
pA = 0.75
pB = 0.25
pBAi = 54/95
pBA2 = 34/90
bayes <- function(a1,a2,a1b,a2b){
    PrAB <- (a1b*a1)/((a1b*a1)+(a2b*a2))
}
```

```r
isoA <- bayes(0.75, 0.25, 54/95, 34/90)
'Probability for someone with isoform A exhibits extreme aggression'
isoA
```

```r
isoB <- bayes(0.25, 0.75,34/90, 54/95)
'Probability for someone with isoform B exhibits extreme aggression'
isoB
```
