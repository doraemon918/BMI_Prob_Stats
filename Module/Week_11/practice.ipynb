```r
data = read.table("NewCFkids.txt",sep = "",stringsAsFactors = F)
colnames(data) = c("ID","FEV1","AGE","GENDER",
                   "PSEUDOA","F508","PANCREAT")

head(data)
```

```r
length(unique(data$ID))
```

```r
data$FEV1 <- as.factor(data$FEV1)
data$ID <- as.factor(data$ID)
data$AGE <- as.factor(data$AGE)
data$GENDER <- as.factor(data$GENDER)
data$F508 <- as.factor(data$F508)
```

```r
apply(data, 2, function(x) length(x[x==-999]))
      
table(data$FEV1)
```

```r
data$FEV1
```

```r
install.packages('mosaic')
library(mosaic)
favstats(data$GENDER)
favstats(data$FEV1)
```

```r
library(lattice)
sam <- as.character(sample(unique(data$ID),10))

sam2 <- data[data$ID %in% sam,]
xyplot(FEV1 ~ GENDER | ID, sam2, type = c("g","p","r","l"),
      index = function(x,y) coef(lm(y ~x))[1],
       aspect = "xy")

xyplot(FEV1 ~ AGE | ID, sam2, type = c("b"),
      index = function(x,y) coef(lm(y ~x))[1],
       aspect = "xy")
```

```r
sam <- as.character(sample(unique(data$ID),10))

sam3 <- data[data$ID %in% sam,]
xyplot(F508 ~ AGE | ID, sam3, type = c("g","p","r","l"),
      index = function(x,y) coef(lm(y ~x))[1],
       aspect = "xy")
```
