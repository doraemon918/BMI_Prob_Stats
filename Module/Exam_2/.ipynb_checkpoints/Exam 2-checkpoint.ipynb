$6$. Consider the following scenario: A data scientist has been asked to conduct a research on the effect of a treatment on anorexia patients. The data corresponded to weight change data for young female anorexia patient. The data contain three columns:


- treat: Factor of three levels: "Cont" (control), "CBT" (Cognitive Behavioural treatment) and "FT" (family treatment).
- Prewt: Weight of patient before study period, in lbs.
- Postwt:Weight of patient after study period, in lbs.

A. Is there a difference between the mean female patiente weights before and after the study across all treatments?

```r
library(MASS)
attach(anorexia)
head(anorexia)
```

```r
"comparision the mean of weight between pre and post treatment "
library(dplyr)
"pretreat"
group_by(anorexia, anorexia$Treat) %>%
  summarise(
    count = n(),
    mean = mean(anorexia$Prewt, na.rm = TRUE),
    sd = sd(anorexia$Prewt, na.rm = TRUE)
  )
"Posttreat"
group_by(anorexia, anorexia$Treat) %>%
  summarise(
    count = n(),
    mean = mean(anorexia$Postwt, na.rm = TRUE),
    sd = sd(anorexia$Postwt, na.rm = TRUE)
  )
```

$10$. The dataset anscombe.txt represent 4 different datasets constructed by Anscombe in 1973. The x and y variables are matched to each dataset respectively (e.g. x1 corresponds to y1, etc). Produce a Simple linear regression for each dataset (write the linear equation for each model), generate residuals plots, and discuss which is the most appropiate model from the 4 datasets and why.

```r
ansc = read.table(file = "anscombe.txt",header = T,stringsAsFactors = F) #need to have header, otherwise the dataset will not work
#colnames(ansc) <- ansc[1,]
#ansc <- ansc[-1,]
ansc
#str(ansc)
```

```r
#install.packages("mosaic")
#install.packages('swirl')
install.packages("calibrate")
library(mosaic)
library(tibble)
library(dplyr)
library(ggplot2)
library(mosaic)
library(swirl)
```

```r
'first need to clean up the data by putting dataset to dataframes'
xy1 <- data.frame(x=ansc[['x1']], y=ansc[['y1']], Set = 'xy1')
xy2 <- data.frame(x=ansc[['x2']], y=ansc[['y2']], Set = 'xy2')
xy3 <- data.frame(x=ansc[['x3']], y=ansc[['y3']], Set = 'xy3')
xy4 <- data.frame(x=ansc[['x4']], y=ansc[['y4']], Set = 'xy4')
data <- rbind(xy1,xy2,xy3,xy4) #combine all dataframe to one 
aggregate(cbind(x,y)~ Set, data, mean) 
aggregate(cbind(x,y)~ Set, data, sd) 
```

```r
m1 <- lm(y~x, data, Set =='xy1')
m2 <- lm(y~x, data, Set =='xy2')
m3 <- lm(y~x, data, Set =='xy3')
m4 <- lm(y~x, data, Set =='xy4')
```

```r
summary(m1)
summary(m2)
summary(m3)
summary(m4)
```

```r
'xy1 - good'
par(mfrow = c(2,2))
plot(m1, main = 'xy1')
```

```r
'xy2 - no'
par(mfrow = c(2,2))
plot(m2, main = 'xy2')
```

```r
'xy3 - good but there"s an outlier'
par(mfrow = c(2,2))
plot(m3, main = 'xy3')
```

```r
'xy4 - doesn"t fit any linear model'
par(mfrow = c(2,2))
plot(m4, main = 'xy4')
```

```r
#ggplot for all dataset
gg <- ggplot(data, aes(x = x, y = y))
gg <- gg + geom_point(color = "black")
gg <- gg + facet_wrap(~Set, ncol = 2)
gg <- gg + geom_smooth(formula = y ~ x, method = "lm", se = FALSE, data = data)
gg
'from all the plots, we can see that dataset 1 is a well fit model compared to the rest'
```
