### A bulb company company produces high efficient bulbs. The probability a bulb to be defective is 0.01. 

a. What is the probability that a sample of 300 bulbs will contain exactly 5 defective?<br><br>
b. Draw the histogram of probabilities

```r
"
mean is 0.01*300 = 3, so the probability of 5 out of 300 is:
"
dpois(5,3)
plot(0:5, dpois(0:5, 3), type = "h", ylim = c(0, 0.25),
    xlab = "X", main = "hist", ylab = "P(X)", lwd = 3, 
    cex.lab = 1.5, cex.axis = 2,
                          cex.main = 2) 
```

### B. X is with mean μ = 40, Find and plot the density distribution with three different standard deviations (2,6,12) (Please draw the curves on the same figure with different colors and a legend) for:

a. P(x < 40)<br><br>
b. P(x > 21)

```r
'a'
pnorm(40,40,2)
pnorm(40,40,6)
pnorm(40,40,12)
a <- rnorm(40,40,2)
b <- rnorm(40,40,6)
c <- rnorm(40,40,12)
da <- density(a)
db <- density(b)
dc <- density(c)
plot(da, xlim =c(-5,80))
lines(db, col = "green")
lines(dc, col = "red")
```

```r
'b'
pnorm(21,40,2, lower.tail = F)
pnorm(21,40,6, lower.tail = F)
pnorm(21,40,12, lower.tail = F)
```

### C. The file plant_heights.csv contains the heights for 1000 plants chosen along a altitudinal gradient (from low to high elevation) in a natural park, the aim of the study is to understand canopy density across a gradient. 

Apply some of the descriptive statistics we have seem so far (histograms, density plots, qqplots) to explain the behavior of the data, write a summary paragraph that gives an initial overview of what you see in this data (are the plants randomly distributed across the gradient? what kind of distribution does the data have and how can you explain it in a biological sense.

hint(try to plot the data using the log transformation, which will make the patterns more evident)

```r
plant = read.table(file = "plant_heights.csv", header = T)
tail(plant)
```

```r
"
histogram from original data:
"
hist(plant$x,xlab = "heights", ylab = "altitude", main = expression(italic("Plant heights")))



"
Because the using the data, the histogram is too skewer, we need to
apply transformation, in order for having better view of the data distribution:
"
plant$log = log(plant$x)
hist(plant$log, xlab = "heights", ylab = "altitude", main = expression(italic("Plant heights log")))
```

```r
par(mfrow = c(1,2))
dplant = density(plant$log)
plot(dplant)

qqnorm(plant$log)
qqline(plant$log)
```
