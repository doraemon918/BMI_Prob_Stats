### B. 

A study looked at various cardiovascular risk factors in children, as measured at birth and during their first five years of life. One of the results for newborns were to study differences in bits per minute from two different races.
Based on the data, Is there evidence that the number of heart beats/min of newborn white children is fewer than that of newborn black children? 

The data is located in this working directory with the name Cardiovascular_children.csv

```r
cardio_child <- read.csv(file = "Cardiovascular_children.csv",
                      header = TRUE,
                      dec = ".")
head(cardio_child)
```

```r
'null hypothesis is the number of heart beat per min of white newborn is fewever than that of black newborn'
cardio_w <- cardio_child[cardio_child$race =='white',]
cardio_b <- cardio_child[cardio_child$race =='black',]
white_c <- cardio_w$bpm
black_c <- cardio_b$bpm

"
var is unequaled and not paired
"
t.test(white_c,black_c, var.equal = F, paired = F)
qt(0.975, 37.589)
"null hypothesis showed it's true"
```

```r
dwhite <- density(white_c)
dblack <- density(black_c)
plot(dwhite)
lines(dblack,col="red")
```

```r
#install.packages("car")
library(car)
Boxplot(bpm ~ race, 
        ylab = "bpm",
        xlab = "race", 
        data= cardio_child,
        main = expression(italic("Relation between bmp and race of newborn")))

stripchart(bpm ~ race, data = cardio_child, 
           vertical = TRUE, method = "jitter", 
           pch = 21, 
           add = TRUE,col=rgb(1, 0, 0,0.5)) 
'heart beat per minute of the black newborn are much higher than white newborn. However, the amount of the data we have for black newborn is insuficient for the accuracy of the statistic'

'there are also a lot of outliers '
```

### C. 

A 1980 study was conducted whose purpose was to compare the indoor air quality in offices where smoking was permitted with that in offices where smoking was not permitted.  Measurements were made of carbon monoxide at 1:20 pm in work areas where smoking was permitted and in areas where smoking was prohibited with the following results:

Does CO vary differently in the two types of working environments?


The data is located in this working directory with the name smoking.csv

```r
smoking <- read.csv(file = "smoking.csv",
                      header = TRUE,
                      dec = ".")
tail(smoking)
```

```r
boxplot(smoking)
'we can see that the air quality of area permitted smoking is having the mean of carbon monoxide level higher than the prohibited area'
```

```r
smoking$log = log(smoking$permitted)
#smoking$log10 = log(bimodalData_s$x,10)
#plot(smoking)
plot(smoking$log)
```

```r
t.test(na.omit(smoking$permitted),smoking$prohibited, var.equal=FALSE, paired=FALSE)
'this is unequal pair test, since there are some null/na values in the permitted column'
't-test showed that the mean of carbon monoxide for permitted is higher than prohibited.'
```

### D. 

A study was conducted to investigate the effect of physical training on the serum cholesterol level.  Thirty subjects participated in the study.  Prior to training, blood samples were taken to determine the cholesterol level of each subject.  Then the subjects were put through a training program that centered on daily running and jogging.  At the end of the training period, blood samples were taken again and a second reading on the serum cholesterol level was obtained. Is there an effect of the training on cholesterol levels?

The data is located in this working directory with the name training.csv

```r
training <- read.csv(file = "training.csv",
                      header = TRUE,
                      dec = ".")
tail(training)
pre <- training$Pre.training.level.x..mg.Dl.
pre
post <- training$Post.training.level.x..mg.Dl.
post
```

```r
'null hypothesis is there is no changes in pre training and post training'
boxplot(pre,post)
'from the boxplot, we can see that the post training has lower mean of cholesterol than the pre training'
'there are some outliers, but they are few enough to affect the outcome of the data.'
```

```r
t.test(pre,post, paired=TRUE)
'this test showed the null hypothesis is false and the alternate is post training is lower than the pre training'
```
